```python
import pandas as pd
```

```python
#Links for each of the three original raw datasets for each state
RaceWashington='https://github.com/NichMaue/Projects/raw/main/WARaceInfo.xlsx' #Washington Link
RaceOregon='https://github.com/NichMaue/Projects/raw/main/ORRaceData.xlsx' #Oregon Link
RaceCalifornia='https://github.com/NichMaue/Projects/raw/main/CARaceData.xlsx'#California Link
```

```python
racewa=pd.read_excel(RaceWashington)
raceor=pd.read_excel(RaceOregon)
raceca=pd.read_excel(RaceCalifornia)
```

First, I want to delete data from years 2010-2018 so that only the most recent demographic information is in the data frame:

```python
racewa.query('Year==2019',inplace=True)
```

Then, I want to eliminate total count by age group to eliminate redundancies in count for each race:

```python
racewa=racewa[racewa['Age Group']!='Total']
```

I want to only keep consecutive age groups, so I will create a function that eliminates non-consecutive groups:

```python
stay=['0-19', '20-64', '65+']

racewa=racewa[racewa['Age Group'].isin(stay)]
```

```python
racewa
```

I also want to get rid of total values for Washington, since we only want county information:

```python
racewa=racewa[racewa["Area Name"]!='Washington'] 
```

Now I will reduce each county to a single row and adds together by the three consecutive age groups:

```python
racewa=racewa.groupby(['Area Name','Area ID','Year']).sum()
racewa
```

We can now get rid of the Age Group column, since we have summed our totals for each county: 

```python
racewa.drop(columns=['Age Group'],inplace=True)
racewa
```

Now, I want to keep only columns that have total values for each race category. First, I create a subset that includes all of these columns. Then, I run a function that will keep this subset:

```python
[name for name in racewa.columns if 'Total' in name]
```

```python
racewa=racewa.loc[:,[name for name in racewa.columns if 'Total' in name]]
racewa
```

Now that I have only totals, I reset the index so that Area Name, Area ID, and Year are now columns within the data frame again:

```python
racewa.reset_index(drop=False,inplace=True)
```

```python
racewa
```

I can now drop columns that I do not need, which include Area ID and Year:

```python
racewa.drop(columns=["Area ID", "Year"],inplace=True)
racewa
```

Finally, having total numbers for each racial group is not helpful for a robust analysis, as it prevents a sound comparison between counties with different populations. I need to calculate each racial category as a percentage of the total population:

```python
import numpy as np
```

First, I will round up all values, since some have a decimal:

```python
westcoastrace.iloc[:,2:8]=westcoastrace.iloc[:,2:8].apply(np.ceil)
```

I want to replace the total population by summing up the values of each racial category:

```python
westcoastrace.iloc[:,2:8].sum(axis=1) # axis=0 will sum by column
```

```python
westcoastrace['total_population']=westcoastrace.iloc[:,2:8].sum(axis=1)
```

I divide each column by total and multiply by 100 to get the share in percentages:

```python
westcoastrace.iloc[:,2:8].div(westcoastrace.total_population, axis=0)*100 
```

Next, I want to create a new data frame with just the percentages for each race category. I also want to change the column names:

```python
# new data frame with percent
racepct=westcoastrace.iloc[:,2:8].div(westcoastrace.total_population, axis=0)*100
```

```python
# change column names
racepct.columns+'_pct'
```

```python
racepct.columns=racepct.columns+'_pct'
```

Check the result:

```python
racepct
```

I only want two decimal places, so I round to the nearest two decimal spots:

```python
racepct.round(2)
```

```python
racepct=racepct.round(2)
```

Lastly, I want to concatenate my original data frame with the race percentages:

```python
pd.concat([westcoastrace,racepct],axis=1)
```

```python
westcoastrace=pd.concat([westcoastrace,racepct],axis=1)
```
